```python
import pandas as pd
```

```python
cnn_features = pd.read_csv('/home/bonnie/Documents/OneDrive_UofT/EVLP_X-ray_Project/evlp_xray_cv/inference/resnet50_CADLab_recipient_all-features.csv')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
```

```python
pca.fit(cnn_features)
```

```python
print(pca.explained_variance_ratio_)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
cnn_features.shape
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
exp_var_pca = pca.explained_variance_ratio_
#
# Cumulative sum of eigenvalues; This will be used to create step plot
# for visualizing the variance explained by each principal component.
#
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
#
# Create the visualization plot
#
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual Explained Variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid',label='Cumulative Explained Variance')
plt.ylabel('Explained Variance Ratio', fontsize=14)
plt.xlabel('Principal Component Index', fontsize=14)
plt.legend(loc='best', fontsize=10)
plt.xticks(range(0,len(exp_var_pca)), fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('/home/bonnie/Documents/OneDrive_UofT/EVLP_X-ray_Project/evlp_xray_cv/inference/Adam_best_models/pca_explained_variance.png', dpi=600)
plt.show()
```
